Resize the images in similar size 

```python
from PIL import Image
import os

thumbnail_size = (150, 150) #resize in pixels

base_dir = 'bttai-ajl-2025/train'  

output_dir = 'bttai-ajl-2025/train'  #new folder for thumbnails

# Create the output directory if it doesn't exist
# os.makedirs(output_dir, exist_ok=True)

# Recursively go through all files in the 'train' directory and its subdirectories
for root, dirs, files in os.walk(base_dir):
    for filename in files:
        if filename.endswith('.jpg'):
            # full path to the original image
            img_path = os.path.join(root, filename)
            
            # Open image using PIL
            with Image.open(img_path) as img:
                
                # Resize image and maintain aspect ratio
                img.thumbnail(thumbnail_size)
                
                # Define the output subdirectory structure to mirror the original directory structure
                relative_path = os.path.relpath(root, base_dir)  
                output_subdir = os.path.join(output_dir, relative_path)  # Corresponding subdir in the output directory
                
                # Create the output subdirectory if it doesn't exist
                # os.makedirs(output_subdir, exist_ok=True)
                
                # Define the resized image path (same filename in the new subdirectory)
                resized_img_path = os.path.join(output_subdir, filename)
                
                # Save the new image 
                img.save(resized_img_path)

            print(f"Thumbnail created and saved: {resized_img_path}")

```

Rotating the images 

```python
import random

# Path to the 'train' directory containing subdirectories (class folders)
base_dir = 'bttai-ajl-2025/train'  # Path to your training directory

# Directory to save rotated images in 'trained_resized' directory
output_dir = 'bttai-ajl-2025/train'  # New folder for rotated images

# Recursively go through all files in the 'train' directory and its subdirectories
for root, dirs, files in os.walk(base_dir):
    for filename in files:
        # Check if the file is an image
        if filename.endswith('.jpg'):
            # Full path to the original image
            img_path = os.path.join(root, filename)
            
            # Open the image using PIL
            with Image.open(img_path) as img:
                # Apply a random rotation between -30 and 30 degrees
                rotation_degree = random.uniform(-30, 30)  # Random rotation degree
                img_rotated = img.rotate(rotation_degree, expand=True)  # Rotate and expand to avoid cropping
                
                # Define the output subdirectory structure to mirror the original directory structure
                relative_path = os.path.relpath(root, base_dir)  # e.g., 'class_a'
                output_subdir = os.path.join(output_dir, relative_path)  # Corresponding subdir in the output directory
                
                # Define the rotated image path (same filename in the new subdirectory)
                rotated_img_path = os.path.join(output_subdir, f"rotated_{rotation_degree:.2f}_{filename}")
                
                # Save the rotated image
                img_rotated.save(rotated_img_path)

            print(f"Rotated image saved: {rotated_img_path}")
```
